## Distribuições de Variáveis Aleatórias Contínuas

Variáveis aleatórias contínuas são variáveis que podem assumir qualquer valor dentro de um intervalo, como a altura de uma pessoa ou a temperatura em uma cidade. Estas variáveis podem estar distribuídas de muitas maneiras que serão demonstradas a seguir. 

Alguns métodos comuns utilizados na biblioteca scipy.stats são os seguintes: 

- cdf(x, loc=0, scale=1): Cumulative distribution function.
- pdf(x, loc=0, scale=1): Probability density function.
- rvs(loc=0, scale=1, size=1, random_state=None): Random variates.

Obs. Não será abordado a distribuição uniforme novamente, visto ter sido exposto em notebook anterior. 

### Distribuição Exponencial

A distribuição exponencial é usada para modelar o tempo entre eventos sucessivos independentes com base em um processo de Poisson.

- λ (parâmetro de taxa): Média de eventos por unidade de tempo.
- θ (parâmetro de escala): Dispersão dos valores.
- loc (parâmetro de localização): Desloca a distribuição no eixo. 

```python
from scipy.stats import expon

# Parâmetro da distribuição Exponencial
lambda_ = 0.5  # Taxa (λ)
scale = 1 / lambda_  # Escala = 1/λ
loc = 0  # Localização

# Valores para o eixo x
x = np.linspace(1, 21, 100)

# Densidade de probabilidade da distribuição Exponencial
pdf = expon.pdf(x, loc=loc, scale=scale)

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'λ: {lambda_}')
plt.title('Distribuição Exponencial')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```

### Distribuição Gama

A distribuição Gama é uma extensão da distribuição exponencial e é frequentemente usada para modelar tempos de espera até a ocorrência de múltiplos eventos. Ela é parametrizada por dois parâmetros:

- k (parâmetro de forma): controla a forma da distribuição.
- θ (parâmetro de escala): afeta a dispersão dos valores.

```python
from scipy.stats import gamma

# Parâmetros da distribuição gama
k = 2.0  # Parâmetro de forma
loc = 0  # Localização
scale = 2.0  # Escala

# Valores para o eixo x
x = np.linspace(1, 21, 100)

# Densidade de probabilidade
pdf = gamma.pdf(x, k, loc, scale)

plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'p: {k}')
plt.title('Distribuição Gama')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.show()
```

### Distribuição Qui-Quadrado

É uma distribuição utilizada em testes de hipóteses e análises de variância.

- k (graus de liberdade): Corresponde ao número de variáveis normais independentes usadas na soma dos quadrados.

```python
from scipy.stats import chi2

# Parâmetros da distribuição qui-quadrado
df = 5 # Graus de Liberdade

# Valores para o eixo x
x = np.linspace(1, 21, 100)

# Densidade de probabilidade
pdf = chi2.pdf(x, df=7)

plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'p: {k}')
plt.title('Distribuição Qui-Quadrado')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.show()
```

### Distribuição Normal

A distribuição normal é uma distribuição de probabilidade contínua que é amplamente utilizada pois muitos fenômenos naturais e sociais diferentes. Ela é caracterizada por sua forma de sino (curva em forma de gaussiana) e simétrica em torno da média.

- μ (média): Representa o valor central da distribuição.
- σ (desvio padrão): Mede a dispersão dos dados em relação à média. Quanto maior o desvio padrão, mais "espalhada" será a distribuição.

```python
from scipy.stats import norm

# Parâmetros da distribuição normal
mu = 0  # Média
sigma = 2  # Desvio padrão

# Valores para o eixo x
x = np.linspace(-10, 10, 100)

# Densidade de probabilidade
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Média: {mu}, Desvio Padrão: {sigma}')
plt.title('Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```

### Distribuição t de Student

A distribuição t de Student é uma distribuição muito utilizada em estatísticas inferenciais. Ela é parecida com a distribuição normal, mas tem caudas mais "gordas", o que permite lidar melhor com amostras pequenas.

```python
from scipy.stats import t

# Parâmetros da distribuição t de Student
df = 5 # Graus de liberdade

# Valores para o eixo x
x = np.linspace(-10, 10, 100)

# Densidade de probabilidade
pdf = t.pdf(x, df)

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Graus de Liberdade: {df}')
plt.title('Distribuição t de Student')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```

### Distribuição F de Snedecor

A distribuição F de Snedecor, também conhecida como distribuição F, é uma distribuição utilizada principalmente em testes de hipóteses e análises de variância (ANOVA). Ela descreve a razão de duas variâncias amostrais, cada uma sendo uma variável aleatória com distribuição qui-quadrado, normalizada por seus graus de liberdade.

- d1 (graus de liberdade do numerador): Refere-se aos graus de liberdade da primeira variável qui-quadrado, geralmente associada ao numerador da razão.
- d2 (graus de liberdade do denominador): Refere-se aos graus de liberdade da segunda variável qui-quadrado, associada ao denominador da razão.

```python
from scipy.stats import f

# Parâmetros da distribuição f de Snedecor
df_num = 5  # graus de liberdade do numerador
df_den = 10  # graus de liberdade do denominador

# Valores para o eixo x
x = np.linspace(1, 10, 100)

# Densidade de probabilidade
pdf = f.pdf(x, df_num, df_den)

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'df_num: {df_num}, df_den: {df_den}')
plt.title('Distribuição f de Snedecor')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```
